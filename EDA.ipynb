```python
diffc = summary_2122 - summary_1819
diffc #analyse individual question later in corr analysis
```

```python
diffc.iloc[:,-2] #change in student satisfaction - 2018-2019 -> 2021-2022
```

```python
print(f"Largest positive change is {max(diffc.iloc[-1,:-1])} for question {diffc.iloc[-1,:-1].idxmax(axis=0)}")
print(f"Largest negative change is {min(diffc.iloc[-1,:-1])} for question {diffc.iloc[-1,:-1].idxmin(axis=0)}")
```

```python
diffc.iloc[-1,:-1] #difference in response, - avg_res
```

```python
#do this to get red and green
x = np.arange(27)
y = diffc.iloc[-1,:-1]
mask1 = y < 0
mask2 = y >= 0

plt.bar(x[mask1], y[mask1], color = 'red')
plt.bar(x[mask2], y[mask2], color = 'green')
plt.xlabel("Questions")
plt.xticks(x, summary_1819.columns[:-1], rotation = 45)
plt.ylabel("Change in agreement % from benchmark")
plt.ylim([-23,23])
plt.show()
```

```python
negatives = len(diffc.iloc[-1,:-1][diffc.iloc[-1,:-1] < 0]) #find all questions that recorded reduced agreeement / all questions
print(f"Out of 27 questions, {negatives} questions recorded reductions in agreement, or {negatives/27:.2f}%.")
print(f"Mean difference: {np.mean(diffc.iloc[-1,:-1])}")
```
